```python
class TS:
    def __init__(self):
        self.ho_ten = ""
        self.toan = 0.0
        self.ly = 0.0
        self.hoa = 0.0

    def nhap_thong_tin(self):
        self.ho_ten = input("Nhap ho ten thi sinh: ")
        self.toan = float(input("Nhap diem Toan: "))
        self.ly = float(input("Nhap diem Ly: "))
        self.hoa = float(input("Nhap diem Hoa: "))

    def tinh_tong_diem(self):
        return self.toan + self.ly + self.hoa

    def in_thong_tin(self):
        print(f"Ho ten: {self.ho_ten}")
        print(f"Diem Toan: {self.toan}")
        print(f"Diem Ly: {self.ly}")
        print(f"Diem Hoa: {self.hoa}")
        print(f"Tong diem: {self.tinh_tong_diem()}")
        print("-------------------------")

danh_sach = []

n = int(input("Nhap so luong thi sinh: "))
for i in range(n):
    print(f"\n--- Nhap thong tin thi sinh thu {i+1} ---")
    ts = TS()
    ts.nhap_thong_tin()
    danh_sach.append(ts)

# Loc thi sinh trung tuyen
diem_chuan = 20
trung_tuyen = [ts for ts in danh_sach if ts.tinh_tong_diem() >= diem_chuan]

# Sap xep diem tu cao xuong thap
trung_tuyen.sort(key=lambda x: x.tinh_tong_diem(), reverse=True)

print("\n===== DANH SACH THI SINH TRUNG TUYEN =====")
if not trung_tuyen:
    print("Khong co thi sinh nao trung tuyen.")
else:
    for ts in trung_tuyen:
        ts.in_thong_tin()
```
