```python
class Date:
    def __init__(self, day=1, month=1, year=2000):
        self.day = day
        self.month = month
        self.year = year

    def display(self):
        print(f"{self.day:02d}/{self.month:02d}/{self.year}")

    def input_date(self, prompt="Nhap ngay (dd mm yyyy): "):
        while True:
            try:
                d, m, y = map(int, input(prompt).split())
                self.day = d
                self.month = m
                self.year = y
                break
            except ValueError:
                print("Nhap sai dinh dang. Vui long nhap lai (vd: 01 01 2000)")
class Employee:
    def __init__(self):
        self.ho_ten = ""
        self.ngay_sinh = Date()
        self.ngay_vao_cong_ty = Date()

    def nhap_thong_tin(self):
        self.ho_ten = input("Nhap ho ten nhan vien: ")
        print("Nhap ngay sinh:")
        self.ngay_sinh.input_date()
        print("Nhap ngay vao cong ty:")
        self.ngay_vao_cong_ty.input_date()

    def in_thong_tin(self):
        print("Ho ten:", self.ho_ten)
        print("Ngay sinh:", end=" ")
        self.ngay_sinh.display()
        print("Ngay vao cong ty:", end=" ")
        self.ngay_vao_cong_ty.display()
        print("-" * 40)
def main():
    danh_sach_nv = []
    n = int(input("Nhap so luong nhan vien: "))
    for i in range(n):
        print(f"\n--- Nhap thong tin nhan vien thu {i+1} ---")
        nv = Employee()
        nv.nhap_thong_tin()
        danh_sach_nv.append(nv)

    print("\n=== DANH SACH NHAN VIEN ===")
    for nv in danh_sach_nv:
        nv.in_thong_tin()
main()
```
