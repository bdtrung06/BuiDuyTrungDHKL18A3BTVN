```python
from xml.dom.minidom import Document

# tạo document mới
doc = Document()

# tạo thẻ gốc <company>
company = doc.createElement("company")
doc.appendChild(company)

# thêm thẻ <name> công ty
company_name = doc.createElement("name")
company_name_text = doc.createTextNode("GeeksForGeeks Company")
company_name.appendChild(company_name_text)
company.appendChild(company_name)

# danh sách nhân viên
staff_list = [
    {"id": "1", "name": "Amar Pandey", "salary": "8.5 LPA"},
    {"id": "2", "name": "Akbhar Khan", "salary": "7.0 LPA"},
    {"id": "3", "name": "Anthony Walter", "salary": "3.2 LPA"}
]

# thêm từng nhân viên
for s in staff_list:
    staff = doc.createElement("staff")
    staff.setAttribute("id", s["id"])

    name = doc.createElement("name")
    name.appendChild(doc.createTextNode(s["name"]))
    staff.appendChild(name)

    salary = doc.createElement("salary")
    salary.appendChild(doc.createTextNode(s["salary"]))
    staff.appendChild(salary)

    company.appendChild(staff)

# lưu file XML
with open("new_sample.xml", "w", encoding="utf-8") as f:
    doc.writexml(f, indent="  ", addindent="  ", newl="\n", encoding="UTF-8")

print("Đã tạo file new_sample.xml thành công!")
```
