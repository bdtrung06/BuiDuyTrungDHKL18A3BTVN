```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity         
        self.data = []                   
        self.top = -1                    

    def __del__(self):
        print("Giai phong bo nho ngan xep")
        del self.data

    def isEmpty(self):
        return self.top == -1

    def isFull(self):
        return self.top == self.capacity - 1

    def push(self, value):
        if self.isFull():
            print("Ngan xep da day, khong the push.")
        else:
            self.data.append(float(value))
            self.top += 1

    def pop(self):
        if self.isEmpty():
            print("Ngan xep rong, khong the pop.")
            return None
        else:
            value = self.data.pop()
            self.top -= 1
            return value
stack = Stack(3)

stack.push(1.5)
stack.push(2.5)
stack.push(3.5)

stack.push(4.5)  

print(stack.pop())  
print(stack.pop())  
print(stack.pop())  

print(stack.pop())  
```
