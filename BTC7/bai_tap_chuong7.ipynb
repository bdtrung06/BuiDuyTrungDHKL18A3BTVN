```python
# bài 7.5
import tkinter as tk
from PIL import Image, ImageTk # Cần cài đặt Pillow bằng lệnh: pip install Pillow

def display_image():
    #  Tạo cửa sổ chính
    root = tk.Tk()
    root.title("Chương trình xem ảnh.")
    root.geometry("600x450")
    root.configure(bg="black") # Đặt màu nền đen như trong hình minh họa

    try:
        #  Mở file ảnh (Đảm bảo file 'Otto.png' nằm cùng thư mục với file code)
        # Nếu muốn nâng cấp đọc .jpg, .bmp... chỉ cần đổi tên file tại đây
        img_open = Image.open("Otto.png")
        
        #  Chuyển đổi ảnh sang định dạng mà Tkinter hiểu được
        img_render = ImageTk.PhotoImage(img_open)

        #  Hiển thị ảnh bằng một Label
        label_img = tk.Label(root, image=img_render, bg="black")
        label_img.image = img_render # Giữ tham chiếu để ảnh không bị mất (garbage collection)
        label_img.pack(expand=True, fill="both")

    except FileNotFoundError:
        # Xử lý trường hợp không tìm thấy file ảnh
        error_label = tk.Label(root, text="Không tìm thấy file 'Otto.png'!", fg="white", bg="black")
        error_label.pack(pady=50)

    #  Chạy chương trình
    root.mainloop()

if __name__ == "__main__":
    display_image()
```

```python
# bài 7.7
import tkinter as tk

def calculate_sum():
    try:
        # Lấy giá trị N từ ô nhập liệu và chuyển sang kiểu số nguyên
        n = int(entry_n.get())
        
        if n < 1:
            label_result.config(text="Vui lòng nhập N >= 1", fg="red")
            return
            
        # Tính tổng từ 1 đến N
        total_sum = sum(range(1, n + 1))
        
        # Tạo chuỗi hiển thị theo định dạng: 1 + 2 + 3 + ... + N = Result
        if n <= 5:
            sequence = " + ".join(map(str, range(1, n + 1)))
        else:
            sequence = f"1 + 2 + 3 + ... + {n}"
            
        label_result.config(text=f"The sum is {sequence} = {total_sum}", fg="black")
        
    except ValueError:
        # Xử lý trường hợp người dùng nhập chữ hoặc để trống
        label_result.config(text="Lỗi: Hãy nhập một số nguyên!", fg="red")

#  Khởi tạo cửa sổ
root = tk.Tk()
root.title("Sum Calculator")
root.geometry("450x200")

#  Các thành phần giao diện
label_prompt = tk.Label(root, text="Enter value of integer N :")
label_prompt.grid(row=0, column=0, padx=10, pady=20, sticky="w")

entry_n = tk.Entry(root, width=25)
entry_n.grid(row=0, column=1, padx=10, pady=20)

# Nhãn hiển thị kết quả
label_result = tk.Label(root, text="", font=("Arial", 10, "italic"))
label_result.grid(row=1, column=1, sticky="w")

#  Nút bấm Validate
btn_validate = tk.Button(root, text="Validate", command=calculate_sum, width=15)
btn_validate.grid(row=2, column=1, pady=10)

#  Chạy chương trình
root.mainloop()
```

```python
# bài 7.8
import tkinter as tk

def find_divisors():
    try:
        # Lấy giá trị N từ Entry và chuyển sang kiểu số nguyên
        n = int(entry_n.get())
        
        if n <= 0:
            label_result.config(text="Vui lòng nhập số nguyên dương!", fg="red")
            return
            
        # Tìm các ước số bằng vòng lặp
        divisors = []
        for i in range(1, n + 1):
            if n % i == 0:
                divisors.append(str(i))
        
        # Chuyển danh sách ước số thành chuỗi để hiển thị
        result_str = "    ".join(divisors)
        label_result.config(text=result_str, fg="black")
        
    except ValueError:
        label_result.config(text="Lỗi: Hãy nhập một số nguyên!", fg="red")

#  Khởi tạo cửa sổ
root = tk.Tk()
root.title("Divisors Finder")
root.geometry("450x200")

# Các thành phần giao diện (Sử dụng Grid để giống trong ảnh)
label_prompt = tk.Label(root, text="Enter the value of N :")
label_prompt.grid(row=0, column=0, padx=20, pady=20, sticky="w")

entry_n = tk.Entry(root, width=30)
entry_n.grid(row=0, column=1, padx=10, pady=20)

label_title_result = tk.Label(root, text="The divisors of N :")
label_title_result.grid(row=1, column=0, padx=20, pady=10, sticky="nw")

# Nhãn hiển thị danh sách các ước số
label_result = tk.Label(root, text="", font=("Arial", 10, "bold"), wraplength=250, justify="left")
label_result.grid(row=1, column=1, padx=10, pady=10, sticky="w")

#  Nút bấm Validate
btn_validate = tk.Button(root, text="Validate", command=find_divisors, width=15)
btn_validate.grid(row=2, column=1, pady=20)

#  Chạy chương trình
root.mainloop()
```

```python
#bài 7.9
import tkinter as tk
from tkinter import ttk  # Thư viện chứa Combobox
import math

def solve():
    try:
        #Lấy giá trị từ các ô nhập liệu
        m = int(entry_m.get())
        n = int(entry_n.get())
        
        #  Lấy chức năng người dùng chọn từ Combobox
        selection = combo_function.get()
        
        if selection == "GCD":
            # Tính Ước số chung lớn nhất
            result = math.gcd(m, n)
            label_result.config(text=f"GCD(m, n) = {result}")
        elif selection == "LCM":
            # Tính Bội số chung nhỏ nhất: LCM(a,b) = |a*b| / GCD(a,b)
            result = abs(m * n) // math.gcd(m, n)
            label_result.config(text=f"LCM(m, n) = {result}")
        else:
            label_result.config(text="Vui lòng chọn chức năng!", fg="red")
            
    except ValueError:
        label_result.config(text="Lỗi: Hãy nhập số nguyên!", fg="red")

# Khởi tạo cửa sổ chính
root = tk.Tk()
root.title("GCD & LCM Calculator")
root.geometry("450x250")

# Giao diện nhập m
tk.Label(root, text="Enter value of m:").grid(row=0, column=0, padx=20, pady=10, sticky="w")
entry_m = tk.Entry(root)
entry_m.grid(row=0, column=1, padx=10, pady=10)

# Giao diện nhập n
tk.Label(root, text="Enter value of n:").grid(row=1, column=0, padx=20, pady=10, sticky="w")
entry_n = tk.Entry(root)
entry_n.grid(row=1, column=1, padx=10, pady=10)

# Giao diện chọn chức năng (Combobox)
tk.Label(root, text="Choose function:").grid(row=2, column=0, padx=20, pady=10, sticky="w")
combo_function = ttk.Combobox(root, values=["GCD", "LCM"], state="readonly")
combo_function.grid(row=2, column=1, padx=10, pady=10)
combo_function.set("GCD") # Giá trị mặc định

# Nhãn hiển thị kết quả
label_result = tk.Label(root, text="", font=("Arial", 10, "bold"), fg="blue")
label_result.grid(row=3, column=1, pady=10, sticky="w")

# Nút bấm tính toán (Giống nút Validate trong hình)
btn_solve = tk.Button(root, text="Calculate", command=solve, width=15)
btn_solve.grid(row=4, column=1, pady=10)

root.mainloop()
```

```python
# bài 7.11
import tkinter as tk
from tkinter import messagebox

def tinh_am_lich():
    try:
        nam_duong = int(entry_year.get())
        
        # Danh sách Can và Chi
        can = ["Canh", "Tân", "Nhâm", "Quý", "Giáp", "Ất", "Bính", "Đinh", "Mậu", "Kỷ"]
        chi = ["Thân", "Dậu", "Tuất", "Hợi", "Tý", "Sửu", "Dần", "Mão", "Thìn", "Tỵ", "Ngọ", "Mùi"]
        
        # Công thức tính Can Chi
        ten_can = can[nam_duong % 10]
        ten_chi = chi[nam_duong % 12]
        nam_am_lich = f"{ten_can} {ten_chi}"
        
        # Hiển thị thông báo kết quả
        # Lưu ý: Đề bài yêu cầu hiển thị hình ảnh con giáp, 
        # trong messagebox mặc định chỉ hiện text. Bạn có thể dùng icon hoặc tạo TopLevel riêng.
        messagebox.showinfo("Kết quả", f"Năm Dương lịch {nam_duong} là năm Âm lịch: {nam_am_lich}")
        
    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng nhập năm là một số nguyên!")

#  Khởi tạo cửa sổ
root = tk.Tk()
root.title("Chuyển đổi năm Âm lịch")
root.geometry("350x150")

#  Giao diện người dùng
tk.Label(root, text="Nhập năm Dương lịch:").pack(pady=10)
entry_year = tk.Entry(root, width=20)
entry_year.pack(pady=5)

#  Nút chuyển đổi
btn_convert = tk.Button(root, text="Chuyển đổi", command=tinh_am_lich)
btn_convert.pack(pady=15)

#  Chạy chương trình
root.mainloop()
```

```python
# bài 7.12
import tkinter as tk

def tinh_bmi():
    try:
        #  Lấy dữ liệu từ ô nhập liệu
        can_nang = float(entry_weight.get())
        chieu_cao = float(entry_height.get())
        
        #  Công thức tính BMI: Cân nặng / (Chiều cao ^ 2)
        # Lưu ý: Chiều cao đơn vị mét (m)
        bmi = can_nang / (chieu_cao ** 2)
        
        #  Phân loại kết quả
        if bmi < 18.5:
            ket_luan = "Gầy"
        elif 18.5 <= bmi < 24.9:
            ket_luan = "Bình thường"
        elif 25 <= bmi < 29.9:
            ket_luan = "Thừa cân"
        else:
            ket_luan = "Béo phì"
            
        #  Hiển thị kết quả lên giao diện
        label_bmi_val.config(text=f"BMI: {bmi:.2f}")
        label_status.config(text=f"Kết luận: {ket_luan}")
        
    except ValueError:
        label_status.config(text="Lỗi: Hãy nhập số hợp lệ!", fg="red")
    except ZeroDivisionError:
        label_status.config(text="Lỗi: Chiều cao phải > 0!", fg="red")

# Khởi tạo cửa sổ
root = tk.Tk()
root.title("BMI Calculator")
root.geometry("350x250")

# Giao diện Cân nặng
tk.Label(root, text="Cân nặng (kg):").pack(pady=5)
entry_weight = tk.Entry(root, justify="center")
entry_weight.pack()

# Giao diện Chiều cao
tk.Label(root, text="Chiều cao (m):").pack(pady=5)
entry_height = tk.Entry(root, justify="center")
entry_height.pack()

# Nút tính toán
btn_calc = tk.Button(root, text="Tính BMI", command=tinh_bmi)
btn_calc.pack(pady=15)

# Hiển thị kết quả
label_bmi_val = tk.Label(root, text="BMI: --", font=("Arial", 10, "bold"))
label_bmi_val.pack()

label_status = tk.Label(root, text="Kết luận: --", font=("Arial", 10))
label_status.pack(pady=5)

root.mainloop()
```

```python
#bai 7.14
import tkinter as tk
from tkinter import messagebox

def tinh_cuoc():
    try:
        #  Lấy dữ liệu đầu vào
        km = float(entry_km.get())
        phut_cho = int(entry_wait.get())
        loai_xe = var_xe.get()
        
        #  Khởi tạo các biến tính toán
        tong_tien = 0
        gia_mo_cua = 12000
        khoang_cach_mo_cua = 0.8
        
        # Tiền chờ: 5 phút đầu miễn phí, từ phút thứ 6 là 750đ/phút
        tien_cho = 0
        if phut_cho > 5:
            tien_cho = (phut_cho - 5) * 750
            
        #  Tính toán dựa trên loại xe
        if loai_xe == 4:
            if km <= khoang_cach_mo_cua:
                tong_tien = gia_mo_cua
            elif km <= 30:
                tong_tien = gia_mo_cua + (km - khoang_cach_mo_cua) * 15300
            else:
                tong_tien = gia_mo_cua + (30 - khoang_cach_mo_cua) * 15300 + (km - 30) * 12100
        else: # Xe 7 chỗ
            if km <= khoang_cach_mo_cua:
                tong_tien = gia_mo_cua
            elif km <= 30:
                tong_tien = gia_mo_cua + (km - khoang_cach_mo_cua) * 16100
            else:
                tong_tien = gia_mo_cua + (30 - khoang_cach_mo_cua) * 16100 + (km - 30) * 13800
        
        tong_thanh_toan = tong_tien + tien_cho
        
        #  Hiển thị kết quả
        label_res_cuoc.config(text=f"Tiền cước: {tong_tien:,.0f} VNĐ")
        label_res_cho.config(text=f"Tiền chờ: {tien_cho:,.0f} VNĐ")
        label_total.config(text=f"TỔNG CỘNG: {tong_thanh_toan:,.0f} VNĐ", fg="red")

    except ValueError:
        messagebox.showerror("Lỗi", "Vui lòng nhập số hợp lệ!")

# Khởi tạo cửa sổ
root = tk.Tk()
root.title("Tính Cước Taxi")
root.geometry("400x450")

# Giao diện chọn loại xe
tk.Label(root, text="CHỌN LOẠI XE", font=("Arial", 12, "bold")).pack(pady=10)
var_xe = tk.IntVar(value=4)
tk.Radiobutton(root, text="Xe 4 chỗ", variable=var_xe, value=4).pack()
tk.Radiobutton(root, text="Xe 7 chỗ", variable=var_xe, value=7).pack()

# Giao diện nhập KM và thời gian chờ
tk.Label(root, text="Quãng đường (km):").pack(pady=5)
entry_km = tk.Entry(root, justify="center")
entry_km.pack()

tk.Label(root, text="Thời gian chờ (phút):").pack(pady=5)
entry_wait = tk.Entry(root, justify="center")
entry_wait.pack()

# Nút tính toán
tk.Button(root, text="TÍNH CƯỚC", command=tinh_cuoc, bg="orange", font=("bold")).pack(pady=20)

# Hiển thị kết quả
label_res_cuoc = tk.Label(root, text="Tiền cước: 0 VNĐ")
label_res_cuoc.pack()
label_res_cho = tk.Label(root, text="Tiền chờ: 0 VNĐ")
label_res_cho.pack()
label_total = tk.Label(root, text="TỔNG CỘNG: 0 VNĐ", font=("Arial", 13, "bold"))
label_total.pack(pady=10)

root.mainloop()
```
