```python
from xml.dom import minidom

# đọc file XML cũ
doc = minidom.parse("new_sample.xml")

# lấy thẻ gốc <company>
company = doc.getElementsByTagName("company")[0]

# tạo nhân viên mới
new_staff = doc.createElement("staff")
new_staff.setAttribute("id", "4")

# thêm tên nhân viên
name = doc.createElement("name")
name.appendChild(doc.createTextNode("Lê Thị D"))
new_staff.appendChild(name)

# thêm lương nhân viên
salary = doc.createElement("salary")
salary.appendChild(doc.createTextNode("5.0 LPA"))
new_staff.appendChild(salary)

# thêm vào công ty
company.appendChild(new_staff)

# lưu file XML cập nhật
with open("new_sample_updated.xml", "w", encoding="utf-8") as f:
    doc.writexml(f, indent="  ", addindent="  ", newl="\n", encoding="UTF-8")

print("Đã thêm nhân viên mới và lưu file new_sample_updated.xml")
```
