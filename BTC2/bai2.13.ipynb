```python
from xml.dom import minidom

# đọc file XML
doc = minidom.parse("xml/students.xml")

# lấy tất cả thẻ <student>
students_nodes = doc.getElementsByTagName("student")

# tạo danh sách sinh viên dưới dạng dict
students_list = []
for student in students_nodes:
    sid = student.getAttribute("id")
    name = student.getElementsByTagName("name")[0].firstChild.nodeValue
    birth = student.getElementsByTagName("birth")[0].firstChild.nodeValue
    class_ = student.getElementsByTagName("class")[0].firstChild.nodeValue
    gender = student.getElementsByTagName("gender")[0].firstChild.nodeValue
    students_list.append({"id": sid, "name": name, "birth": birth, "class": class_, "gender": gender})

# sắp xếp theo tên (alphabet)
students_list_sorted = sorted(students_list, key=lambda x: x["name"])

# in kết quả
print("Danh sách sinh viên theo tên (tăng dần):")
for s in students_list_sorted:
    print(f"{s['id']} - {s['name']} - {s['birth']} - {s['class']} - {s['gender']}")
```
