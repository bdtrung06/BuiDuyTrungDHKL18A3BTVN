```python
from xml.dom import minidom

# đọc file XML
doc = minidom.parse("new_sample_updated.xml")

# lấy tất cả thẻ <staff>
staffs = doc.getElementsByTagName("staff")

# duyệt và xóa nhân viên id=2
for staff in staffs:
    if staff.getAttribute("id") == "2":
        staff.parentNode.removeChild(staff)
        break  # dừng sau khi xóa

# lưu file XML mới
with open("new_sample_deleted.xml", "w", encoding="utf-8") as f:
    doc.writexml(f, indent="  ", addindent="  ", newl="\n", encoding="UTF-8")

print("Đã xóa nhân viên id=2 và lưu file new_sample_deleted.xml")
```
