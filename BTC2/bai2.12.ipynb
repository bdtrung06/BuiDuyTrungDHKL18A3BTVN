```python
from xml.dom import minidom

# đọc file XML
doc = minidom.parse("new_sample_deleted.xml")

# lấy tất cả thẻ <staff>
staffs = doc.getElementsByTagName("staff")

# tìm nhân viên id=4 và cập nhật thông tin
for staff in staffs:
    if staff.getAttribute("id") == "4":
        # cập nhật tên
        name_tag = staff.getElementsByTagName("name")[0]
        name_tag.firstChild.nodeValue = "Lê Thị Dương"
        
        # cập nhật lương
        salary_tag = staff.getElementsByTagName("salary")[0]
        salary_tag.firstChild.nodeValue = "5.5 LPA"
        break  # chỉ cập nhật 1 nhân viên

# lưu file XML mới
with open("new_sample_updated_2.xml", "w", encoding="utf-8") as f:
    doc.writexml(f, indent="  ", addindent="  ", newl="\n", encoding="UTF-8")

print("Đã cập nhật thông tin nhân viên id=4 và lưu file new_sample_updated_2.xml")
```
