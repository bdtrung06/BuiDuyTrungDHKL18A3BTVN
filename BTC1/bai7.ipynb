```python
class Date:
    def __init__(self, day=1, month=1, year=2000):
        self.day = day
        self.month = month
        self.year = year

    def display(self):
        print(f"{self.day:02d}/{self.month:02d}/{self.year}")

    def is_leap_year(self):
        return (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0)

    def days_in_month(self):
        if self.month == 2:
            return 29 if self.is_leap_year() else 28
        elif self.month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        else:
            return 30

    def next(self):
        self.day += 1
        if self.day > self.days_in_month():
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1

d = Date(2, 10, 2025)
print("Hom nay:")
d.display()

d.next()
print("Ngay mai:")
d.display()
```
