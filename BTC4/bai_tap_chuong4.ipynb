```python
# bài 4.1
import pandas as pd
import numpy as np
mylist = list('abcdefghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))
ser_from_list = pd.Series(mylist)# từ list
ser_from_array = pd.Series(myarr) # từ array
ser_from_dict = pd.Series(mydict) # từ dict
print("Series từ list:")
print(ser_from_list.head())
print("\nSeries từ numpy array:")
print(ser_from_array.head())
print("\nSeries từ dictionary:")
print(ser_from_dict.head())
```

```python
# bài 4.2
mylist = list('abcdefghijklmnopqrstuvwxyz')
myarr = np.arange(26)
mydict = dict(zip(mylist, myarr))
ser = pd.Series(mydict)
# chuyen doi chuoi ser thanh thanh dataframe va reset index
df = ser.to_frame().reset_index()
df.columns = ['letter', 'number']  # đặt tên cột
print("\nDataFrame sau khi chuyển từ Series và reset index:")
print(df.head())
```

```python
# bài 4.3
import numpy as np
ser1 = pd.Series(list('abcdefghijklmnopqrstuvwxyz'))
ser2 = pd.Series(np.arange(26))
# ket noi series pandas ser1 va ser2 thanh dataframe
df_combined = pd.concat([ser1, ser2], axis=1)
df_combined.columns = ['letter', 'number']  # đặt tên cột
print("\nDataFrame kết hợp từ hai Series:")
print(df_combined.head())
```

```python
# bài 4.4
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
# tu ser1 xoa cac muc co trong ser2
ser_difference = ser1[~ser1.isin(ser2)]
print("\nSeries sau khi loại bỏ các mục có trong ser2 từ ser1:")
print(ser_difference)
# lay tat ca cac muc co trong ser1 va ser2 nhung khong nam trong ca hai
ser_symmetric_difference = pd.concat([ser1[~ser1.isin(ser2)], ser2[~ser2.isin(ser1)]])
print("\nSeries chứa tất cả các mục không nằm trong cả hai Series:")
print(ser_symmetric_difference)
```

```python
# bài 4.5
ser = pd.Series(np.random.normal(10, 5, 25))
# gia tri toi thieu (minimum) cua ser
min_value = ser.min()
print("\nGiá trị tối thiểu của Series:")
print(min_value)
# phan centile thu 25 (25th percentile) cua ser
percentile_25 = ser.quantile(0.25)
print("\nPhần centile thứ 25 của Series:")
print(percentile_25)
# trung vi (median) cua ser
median_value = ser.median()
print("\nTrung vị của Series:")
print(median_value)
# phan centile thu 75 (75th percentile) cua ser
percentile_75 = ser.quantile(0.75)
print("\nPhần centile thứ 75 của Series:")
print(percentile_75)
# gia tri toi da (maximum) cua ser
max_value = ser.max()
print("\nGiá trị tối đa của Series:")
print(max_value)
```

```python
# bài 4.6
ser = pd.Series(np.take(list('abcdefgh'), np.random.randint(8, size=30)))
#1. giai thich y nghia doan code :
# - list('abcdefgh'): tạo list ['a','b','c','d','e','f','g','h']
# - np.random.randint(8, size=30): tạo 30 số ngẫu nhiên từ 0-7
# - np.take(): lấy các phần tử từ list dựa trên chỉ số ngẫu nhiên
# - pd.Series(): tạo Series từ kết quả
# 2. dem so lan xuat hien cua moi gia tri trong ser
value_counts = ser.value_counts()
print("\nSố lần xuất hiện của mỗi giá trị trong Series:")
print(value_counts)
```

```python
# bài 4.7
np.random.RandomState(100)
ser = pd.Series(np.random.randint(1, 5, [12]))
# tim 2 phantu xuat hien nhieu nhat trong ser
top_2 = ser.value_counts().nlargest(2)  
print("\nHai phần tử xuất hiện nhiều nhất trong Series:")
print(top_2)
# thay cac phan tu khong phai 2 phan tu tren bang gia tri other
ser_replaced = ser.apply(lambda x: x if x in top_2.index else 'Other')
print("\nSeries sau khi thay các phần tử không phải hai phần tử trên bằng 'Other':")
print(ser_replaced)
```

```python
# bài 4.8
ser = pd.Series(np.random.random(20))
# chia ser thanh 10 phan vi bang nhau
bins = pd.qcut(ser, 10) 
print("\nCác phần bin sau khi chia Series thành 10 phần vị bằng nhau:")
print(bins)
# thay cac gia tri trong ser bang ten phan vi tuong ung
ser_binned = pd.qcut(ser, 10, labels=[f'Bin{i+1}' for i in range(10)])
print("\nSeries sau khi thay các giá trị bằng tên phần vị tương ứng:")
print(ser_binned)
```

```python
# bài 4.9
ser = pd.Series(np.random.randint(1, 10, 35))
# 1. Giải thích ý nghĩa:
# - np.random.randint(1, 10, 35): tạo 35 số ngẫu nhiên từ 1 đến 9
# - pd.Series(): tạo Series từ mảng numpy
# 2. dinh hinh ser thanh dataframe 7x5
df_reshaped = ser.values.reshape(7, 5)
df_final = pd.DataFrame(df_reshaped, columns=[f'Col{i+1}' for i in range(5)])
print("\nDataFrame sau khi định hình Series thành 7x5:")
print(df_final)
```

```python
# bài 4.10
ser = pd.Series(np.random.randint(1, 10, 7))
print("Series:", ser.values)
# vi tri cac so la boi cua 3 tu ser
# Tìm vị trí các số là bội của 3
positions = np.where(ser % 3 == 0)[0]
values = ser[ser % 3 == 0]
print("Các số là bội của 3:", values.values)
print("Vị trí tương ứng:", positions)
```

```python
# bai 4.12
ser1= pd.Series(range(5))
ser2= pd.Series(list('abcde'))
# tao dataframe tu xep chong ser1 va ser2 theo chieu doc
df_vertical = pd.concat([ser1, ser2], axis=0)
print("\nXếp chồng theo chiều dọc:")
print(df_vertical)
# tao dataframe tu xep chong ser1 va ser2 theo chieu ngang
df_horizontal = pd.DataFrame([ser1, ser2]).T
df_horizontal.columns = ['Numbers', 'Letters']
print("\nDataFrame từ hai Series xếp chồng theo chiều ngang:")
print(df_horizontal)
```

```python
# bài 4.13
ser1 = pd.Series([10, 9, 6, 5, 3, 1 ,12, 8, 13])
ser2 = pd.Series([1, 3, 10, 13])
# lay  vi tri cua cac muc cua ser 2 trong ser1 duoi dang list
positions = [ser1[ser1 == val].index[0] for val in ser2 if val in ser1.values]
print("Vị trí của các mục của ser2 trong ser1 là:", positions)
```

```python
# bài 4.14
truth= pd.Series(range(10))
pred = pd.Series(range(10)) + np.random.random(10)
# sai so binh phuong trung binh cua truth va pred
mse = ((truth - pred) ** 2).mean()
print("Sai số bình phương trung bình (MSE) giữa truth và pred là:", mse)
```

```python
# bài 4.16
import pandas as pd

ser = pd.Series([1, 3, 6, 10, 15, 21, 27, 35])

# Tính hiệu số của các giá trị liên tiếp (first difference)
diff_ser = ser.diff().iloc[1:]   # bỏ phần tử NaN đầu

print("Series gốc:", ser.values)
print("Hiệu số liên tiếp:", diff_ser.values)
# hieu so cua hieu so lien tiep
second_diff_ser = diff_ser.diff().iloc[1:]  # bỏ phần tử NaN đầu
print("Hiệu số của hiệu số liên tiếp:", second_diff_ser.values)
```
