```python
from xml.dom import minidom

# đọc file XML
doc = minidom.parse("xml/sample.xml")

# lấy toàn bộ thẻ <name> nằm trong <staff>
names = doc.getElementsByTagName("staff")

print("Danh sách tên nhân viên:")
for staff in names:
    name = staff.getElementsByTagName("name")[0].firstChild.nodeValue
    print("-", name)
```
