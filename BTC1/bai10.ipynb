```python
class Diem:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def nhap(self):
        self.x = float(input("Nhap hoanh do x: "))
        self.y = float(input("Nhap tung do y: "))

    def hien_thi(self):
        print(f"Toa do tam: ({self.x}, {self.y})")
import math

class Elip(Diem):
    def __init__(self, x=0, y=0, a=1, b=1):
        super().__init__(x, y)
        self.a = a
        self.b = b

    def nhap(self):
        super().nhap()
        self.a = float(input("Nhap ban truc lon a: "))
        self.b = float(input("Nhap ban truc nho b: "))

    def dien_tich(self):
        return math.pi * self.a * self.b

    def hien_thi(self):
        super().hien_thi()
        print(f"Ban truc lon a = {self.a}")
        print(f"Ban truc nho b = {self.b}")
        print(f"Dien tich Elip = {self.dien_tich():.2f}")
class DuongTron(Elip):
    def __init__(self, x=0, y=0, r=1):
        super().__init__(x, y, r, r)
        self.r = r

    def nhap(self):
        super().nhap()
        self.r = float(input("Nhap ban kinh duong tron: "))
        self.a = self.b = self.r

    def dien_tich(self):
        return math.pi * self.r ** 2

    def hien_thi(self):
        super().hien_thi()
        print(f"Ban kinh r = {self.r}")
        print(f"Dien tich duong tron = {self.dien_tich():.2f}")
def main():
    print("=== NHAP ELIP ===")
    elip = Elip()
    elip.nhap()
    print("\nThong tin Elip:")
    elip.hien_thi()

    print("\n=== NHAP DUONG TRON ===")
    dt = DuongTron()
    dt.nhap()
    print("\nThong tin Duong Tron:")
    dt.hien_thi()
main()
```
