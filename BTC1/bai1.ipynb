```python
class HinhChuNhat:
    def __init__(self):
        self.dai = 0
        self.rong = 0

    def nhap_du_lieu(self):
        self.dai = float(input("nhap chieu dai: "))
        self.rong = float(input("nhap chieu rong: "))

    def tinh_chu_vi(self):
        return 2 * (self.dai + self.rong)

    def tinh_dien_tich(self):
        return self.dai * self.rong

    def in_thong_tin(self):
        print("thong tin hinh chu nhat:")
        print("chieu dai",self.dai)
        print("chieu rong",self.rong)
        print("chu vi",self.tinh_chu_vi())
        print("dien tich",self.tinh_dien_tich())

hcn = HinhChuNhat()
hcn.nhap_du_lieu()
hcn.in_thong_tin()
```
