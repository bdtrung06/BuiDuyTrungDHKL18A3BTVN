```python
# bài 6.1
import threading
import time

# Hàm chạy trong mỗi luồng
def worker():
    current_thread = threading.current_thread()
    print("Tên luồng đang chạy:", current_thread.name)
    time.sleep(1)

# Tạo danh sách các luồng
threads = []

for i in range(5):
    t = threading.Thread(target=worker, name=f"Thread-{i+1}")
    threads.append(t)
    t.start()

# Đợi tất cả các luồng chạy xong
for t in threads:
    t.join()

print("Tất cả các luồng đã hoàn thành")
```

```python
#bài 6.2
import threading
import urllib.request
import os

# Hàm tải file
def download_file(url):
    filename = url.split("/")[-1]
    print(f"Bắt đầu tải: {filename}")

    urllib.request.urlretrieve(url, filename)

    print(f"Hoàn thành: {filename}")

# Danh sách URL cần tải
urls = [
    "https://www.w3.org/TR/PNG/iso_8859-1.txt",
    "https://www.w3.org/TR/PNG/example.png",
    "https://www.w3.org/TR/PNG/ai.png"
]

threads = []

# Tạo và chạy các luồng
for url in urls:
    t = threading.Thread(target=download_file, args=(url,))
    threads.append(t)
    t.start()

# Đợi tất cả luồng hoàn thành
for t in threads:
    t.join()

print("Đã tải xong tất cả các tệp")
```

```python
# bài 6.3
import threading

# Luồng in số chẵn
def print_even():
    print("Các số chẵn từ 30 đến 50:")
    for i in range(30, 51):
        if i % 2 == 0:
            print(i)

# Luồng in số lẻ
def print_odd():
    print("Các số lẻ từ 30 đến 50:")
    for i in range(30, 51):
        if i % 2 != 0:
            print(i)

# Tạo hai luồng
thread_even = threading.Thread(target=print_even, name="EvenThread")
thread_odd = threading.Thread(target=print_odd, name="OddThread")

# Khởi động luồng
thread_even.start()
thread_odd.start()

# Đợi hai luồng hoàn thành
thread_even.join()
thread_odd.join()

print("Hoàn thành chương trình")
```

```python
# bài 6.4
import threading
import math

# Danh sách lưu kết quả từng luồng
results = []

# Hàm tính tích trong một khoảng
def partial_factorial(start, end):
    result = 1
    for i in range(start, end + 1):
        result *= i
    results.append(result)

# Số cần tính giai thừa
n = 10

# Tạo hai luồng
t1 = threading.Thread(target=partial_factorial, args=(1, n//2))
t2 = threading.Thread(target=partial_factorial, args=(n//2 + 1, n))

# Khởi động luồng
t1.start()
t2.start()

# Đợi hai luồng hoàn thành
t1.join()
t2.join()

# Nhân kết quả của các luồng
factorial = math.prod(results)

print(f"Giai thừa của {n} là: {factorial}")
```

```python
# bài 6.5
import threading
import requests

# Hàm gửi HTTP request
def fetch_url(url):
    print(f"Bắt đầu request: {url}")
    response = requests.get(url)
    print(f"Hoàn thành: {url} | Status code: {response.status_code}")

# Danh sách các URL cần request
urls = [
    "https://httpbin.org/get",
    "https://jsonplaceholder.typicode.com/posts/1",
    "https://jsonplaceholder.typicode.com/users/1",
    "https://api.github.com"
]

threads = []

# Tạo và chạy các luồng
for url in urls:
    t = threading.Thread(target=fetch_url, args=(url,))
    threads.append(t)
    t.start()

# Đợi tất cả luồng hoàn thành
for t in threads:
    t.join()

print("Đã hoàn thành tất cả HTTP requests")
```

```python
# bài 6.6
import threading
import time

limit = 20
lock = threading.Lock()

# Luồng in số chẵn
def print_even():
    for i in range(0, limit + 1, 2):
        with lock:
            print("Số chẵn:", i)
        time.sleep(0.1)

# Luồng in số lẻ
def print_odd():
    for i in range(1, limit + 1, 2):
        with lock:
            print("Số lẻ:", i)
        time.sleep(0.1)

# Tạo hai luồng
t_even = threading.Thread(target=print_even, name="EvenThread")
t_odd = threading.Thread(target=print_odd, name="OddThread")

# Chạy hai luồng
t_even.start()
t_odd.start()

# Đợi hoàn thành
t_even.join()
t_odd.join()

print("Hoàn thành chương trình")
```

```python
# 6.7
import threading
import time

# Hàm chạy trong mỗi thread
def print_time(thread_name, delay, count):
    print(f"Starting {thread_name}")
    for i in range(count):
        time.sleep(delay)
        print(f"{thread_name}: Web {time.ctime(time.time())}")
    print(f"Exiting {thread_name}")

# Main program
print("Starting Main Thread")

# Tạo các thread
google = threading.Thread(target=print_time, args=("Google", 1, 4))
yahoo = threading.Thread(target=print_time, args=("Yahoo", 2, 4))
facebook = threading.Thread(target=print_time, args=("Facebook", 3, 4))

# Chạy thread
google.start()
yahoo.start()
facebook.start()

# Chờ các thread hoàn thành
google.join()
yahoo.join()
facebook.join()

print("Exiting Main Thread")
```

```python
# bài 6.8
import threading
import random

# Hàm tính tổng list con
def sum_sublist(sublist, results, index):
    print(f"Thread {index + 1} xử lý:", sublist)
    results[index] = sum(sublist)

# ================== MAIN ==================
n = 40                 # số phần tử
num_threads = 4        # số luồng

# Tạo list ngẫu nhiên
lst = [random.randint(0, 10) for _ in range(n)]
print("List:", lst)

# Chia list thành các phần bằng nhau
size = n // num_threads
sublists = [lst[i*size:(i+1)*size] for i in range(num_threads)]

# Mảng lưu kết quả
results = [0] * num_threads
threads = []

# Tạo và chạy thread
for i in range(num_threads):
    t = threading.Thread(
        target=sum_sublist,
        args=(sublists[i], results, i)
    )
    threads.append(t)
    t.start()

# Chờ các thread hoàn thành
for t in threads:
    t.join()

# Tính tổng cuối cùng
total_sum = sum(results)
print("Tổng list:", total_sum)
```

```python
# bài 6.9
import threading
import time

# Hàm chạy cho mỗi thread
def web_task(name, delay, times):
    print(f"Starting {name}")
    for _ in range(times):
        time.sleep(delay)
        print(f"{name}: Web {time.ctime()}")
    print(f"Exiting {name}")

# ================= MAIN =================
print("Starting Main Thread")

# Tạo các thread
t1 = threading.Thread(target=web_task, args=("Google", 1, 4))
t2 = threading.Thread(target=web_task, args=("Yahoo", 2, 4))
t3 = threading.Thread(target=web_task, args=("Facebook", 3, 4))

# Chạy các thread
t1.start()
t2.start()
t3.start()

# Chờ các thread kết thúc
t1.join()
t2.join()
t3.join()

print("Exiting Main Thread")
```

```python
# bài 6.10
import threading
import random

# Hàm tìm max của list con
def find_max(sublist, results, index):
    print(f"Thread {index + 1} xử lý:", sublist)
    results[index] = max(sublist)

# ================= MAIN =================
n = 40                # số phần tử
num_threads = 4       # số luồng

# Tạo list ngẫu nhiên
lst = [random.randint(0, 100) for _ in range(n)]
print("List:", lst)

# Chia list thành các list con
size = n // num_threads
sublists = [lst[i*size:(i+1)*size] for i in range(num_threads)]

# Mảng lưu kết quả max của mỗi thread
results = [0] * num_threads
threads = []

# Tạo và chạy các thread
for i in range(num_threads):
    t = threading.Thread(
        target=find_max,
        args=(sublists[i], results, i)
    )
    threads.append(t)
    t.start()

# Chờ các thread hoàn thành
for t in threads:
    t.join()

# Tìm max cuối cùng
max_value = max(results)
print("Giá trị lớn nhất trong list:", max_value)
```
