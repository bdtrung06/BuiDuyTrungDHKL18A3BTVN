```python
from xml.dom import minidom

# đọc file XML
doc = minidom.parse("xml/sample.xml")

# lấy danh sách tất cả <staff>
staffs = doc.getElementsByTagName("staff")

# tìm staff có id="2" và thay đổi salary
for staff in staffs:
    if staff.getAttribute("id") == "2":
        salary_tag = staff.getElementsByTagName("salary")[0]
        salary_tag.firstChild.nodeValue = "7.0 LPA"

# lưu file XML mới
with open("sample_updated.xml", "w", encoding="utf-8") as f:
    doc.writexml(f, indent="  ", addindent="  ", newl="\n", encoding="UTF-8")

print("Đã cập nhật salary của nhân viên id=2 thành 7.0 LPA và lưu vào sample_updated.xml")
```
