```python
import numpy as np

arr_zeros = np.zeros(10, dtype=int)
arr_zeros[5] = 1
print("Câu 1:", arr_zeros)

arr_h = np.arange(10, 25)
print("Câu 2:", arr_h[::-1])

arr_k = np.array([1, 2, 0, 8, 2, 0, 1, 3, 0, 5, 0])
arr_1 = arr_k[arr_k != 0]
print("Câu 3:", arr_1)

arr_2 = np.append(arr_1, [10, 20])
print("Câu 4:", arr_2)

arr_3 = np.insert(arr_2, 5, 100)
print("Câu 5:", arr_3)

arr_final = np.delete(arr_3, [0, 1, 2])
print("Câu 6:", arr_final)
```
